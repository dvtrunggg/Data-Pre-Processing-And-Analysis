# Câu 1: Pew Research Center 
Cho dữ liệu: pew-raw.csv. Bộ dữ liệu này khám phá mối quan hệ giữa thu nhập và tôn giáo (income & religion) \
Đọc dữ liệu \
Xem xét vấn đề về dữ liệu cần khắc phục \
Chuẩn lại dữ liệu để khắc phục vấn đề trên

```python
import pandas as pd
import numpy as np

%load_ext nb_black
```

```python
df1 = pd.read_csv("pew-raw.csv")
df1.head()
```

```python
"""
Nhận xét: 
tên cột chứa giá trị thay vì chứa biến
trong từng cell chứa tần suất
"""
```

```python
# chuyển sang tidy data
df1_melt = df1.melt(id_vars=["religion"], var_name="income", value_name="freq")
df1_melt.head()
```

```python
"""
cột income chứa giá trị chưa dc chuẩn hóa
"""
```

```python
df1_melt.income = df1_melt.income.str.extract("(\d+)").astype("int")
```

```python
df1_melt.income = df1_melt.income.apply(lambda x: x * 1000)
```

```python
df1_melt.head()
```

# Câu 2: Billboard Top 100 
Cho dữ liệu billboard.csv. Bộ dữ liệu này đại diện cho thứ hạng hàng tuần của các bài hát kể từ thời điểm chúng lọt vào Top 100 của Billboard cho đến 75 tuần tiếp theo. \
Đọc dữ liệu \
Xem xét vấn đề về dữ liệu cần khắc phục \
Chuẩn lại dữ liệu để khắc phục vấn dề trên \
Sau khi chuẩn lại dữ liệu, có điều gì phát sinh cần khắc phục tiếp theo không? Nếu có thì hãy chuẩn lại dữ liệu mới có ở câu trên \

```python
df2 = pd.read_csv("billboard.csv", encoding="mac_latin2")
df2.head()
```

```python
"""
Nhận xét:
Tên cột chứa giá trị thay vì chứa biến (x1st, ...) trong từng cell thứ hạng
nếu 1 bài hát nằm trong top 100 dưới 75 lần thì các cột còn lại chứa giá trị NaN

"""
```

```python
df2_melt = df2.melt(
    id_vars=[
        "year",
        "artist.inverted",
        "track",
        "time",
        "genre",
        "date.entered",
        "date.peaked",
    ],
    var_name="week",
    value_name="rank",
)
```

```python
df2_melt.head()
```

```python
df2_melt.tail()
```

```python
"""
chuẩn hóa dữ liệu: 
thay chuỗi thành số trong week
xử lý các dữ liệu NaN
chuyển values của rank thành số (nếu nó là kiểu != numeric)
"""
```

### Thay chuỗi thành số trong week


```python
df2_melt.week = df2_melt.week.str.extract("(\d+)", expand=False).astype(int)
```

```python
df2_melt
```

### Xóa các dòng có giá trị null

```python
df2_melt.dropna(inplace=True)
df2_melt
```

```python
df2_melt["rank"] = df2_melt["rank"].astype(int)
```

```python
df2_melt.head()
```

```python
"""
vấn đề phát sinh: 
nhiều record (song & rank)trong cùng 1 bảng duy nhất

kiểu như vi phạm dạng chuẩn á

=> xử lý:
tạo 1 df songs chứa thông tin của nó
tạo 1 df rank chứa thông tin của rank song, ...
"""

```

```python
df2_song = df2_melt[["year", "artist.inverted", "track", "time", "genre"]]
df2_song = df_song.drop_duplicates()
df2_song = df_song.reset_index()
df2_song["id"] = df_song.index()
```

```python
df2_song
```

```python
df2_rank = pd.merge(
    df2_melt, df2_song, on=["year", "artist.inverted", "track", "time", "genre"]
)
df2_rank = df2_rank[["date.entered", "date.peaked", "week", "rank", "id"]]
df2_rank.rename(columns={"id": "song_id"}, inplace=True)
```

```python
df2_rank.head(10)
```

```python
# tạo cột date
df2_rank["date"] = (
    pd.to_datetime(df2_rank["date.entered"])
    + pd.to_timedelta(df2_rank["week"], unit="w")
    - pd.DateOffset(weeks=1)
)
```

```python
df2_rank.head()
```

```python
df2_rank = df2_rank.drop(["date.entered", "date.peaked"], axis=1)
```
