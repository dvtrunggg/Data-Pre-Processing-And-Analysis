```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
df = pd.DataFrame({
    'Name': ['Lucy', 'John', 'Tom', 'Mary', 'Mark'],
    'PlaceOfBirth': ['NewJork', 'London', 'HaNoi', 'BinhDuong', 'Dubai'],
    'Age': [24, 29, 21, 30, 27]
})
df
```

```python
df_new1 = pd.get_dummies(data=df, columns=['PlaceOfBirth'], prefix='place')
df_new1
```

```python
df_new2 = pd.get_dummies(data=df, columns=['PlaceOfBirth'], prefix='place', drop_first=True)
df_new2
```

```python
df2 = pd.DataFrame({
    'Name': ['Lucy', 'John', 'Tom', 'Mary', 'Mark'],
    'Sex': ['Female', 'Male', 'Other', 'Female', 'Male'],
    'PlaceOfBirth': ['NewJork', 'London', 'HaNoi', 'BinhDuong', 'Dubai'],
    'Age': [24, 29, 21, 30, 27]
})
df2
```

```python
df_new3 = pd.get_dummies(data=df2, columns=['Sex', 'PlaceOfBirth'], prefix=['S', 'P'], drop_first=True)
df_new3
```
