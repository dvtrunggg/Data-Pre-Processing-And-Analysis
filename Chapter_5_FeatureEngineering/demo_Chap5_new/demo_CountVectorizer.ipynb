```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/inaugural_speeches_clean.csv')
```

```python
df[['text_clean', 'char_count', 'word_count', 'avg_word_length']].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# khởi tạo countVectorizer
cv = CountVectorizer()
```

```python
# phân rã nội dung văn bản
# cv.fit(df['text_clean'])
# df_transform = cv.transform(df['text_clean'])
df_transform = cv.fit_transform(df['text_clean'])
```

```python
print(cv.get_feature_names())
```

```python
len(cv.get_feature_names())
```

```python
df_transform = df_transform.toarray()
```

```python
df_transform
```

```python
# chuyển về DataFrame
df_transform = pd.DataFrame(df_transform, columns = cv.get_feature_names()).add_prefix('counts_')
df_transform.head()
```

```python
df_transform.info()
```

```python
df_new = pd.concat([df, df_transform], axis=1)
df_new.drop(['text','text_clean'], axis=1, inplace=True)
df_new.head()
```

```python
df_new.info()
```
