```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame({
    'Gender': ['M', 'F', 'M', 'M', 'F', 'M', np.nan, 'M'],
    'Manpower': [25, np.nan, 32, np.nan, 26, 29, 24, 32],
    'Sales': [344, 270, 320, 270, np.nan, 315, 262, 290]
})
df
```

```python
df_no_missing = df.dropna()
df_no_missing
```

```python
df_no_missing = df.dropna(axis=1)
df_no_missing
```

```python
#Manpower -> thay thế bằng giá trị tương tự
# tính mean của M -> thay cho M
# tính mean của F -> thay cho F
mean_m = df.loc[df['Gender'] == 'M', 'Manpower'].mean()
mean_m
```

```python
mean_f = df.loc[df['Gender'] == 'F', 'Manpower'].mean()
mean_f
```

```python
df.loc[(df['Gender'] == 'M') & (df['Manpower'].isnull()), 'Manpower'] = mean_m
df.loc[(df['Gender'] == 'F') & (df['Manpower'].isnull()), 'Manpower'] = mean_f
```

```python
df
```

```python
# Sales --> thay bằng giá trị tổng quát
mean_sale = df['Sales'].mean()
df.loc[df['Sales'].isnull(), 'Sales'] = mean_sale
df
```
