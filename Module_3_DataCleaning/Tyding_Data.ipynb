```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'hoten': ['nva', 'nvb', 'nvc'],
    'toan': [7, 6, 8],
    'ly': [5.5, 8, 6],
    'hoa': [9, 7, 8]
})
df
```

#### Vấn đề gặp phải 

**Cột chứ giá trị thay vì chứa biến**

```python
# sử dụng pd.melt để chuyển data từ dạng thống kê sang data dạng phân tích
df_new1 = pd.melt(frame=df, id_vars=['hoten'], value_vars=['toan', 'ly', 'hoa'])
df_new1
```

```python
df_new2 = pd.melt(frame=df, id_vars=['hoten'], value_vars=['toan', 'ly', 'hoa'], 
                 var_name='mon', value_name='diem')
df_new2
```

```python
df_new3 = pd.melt(frame=df, id_vars=['hoten'], var_name='mon', value_name='diem')
df_new3
```

## Pivot data

```python
df_pivot = df_new3.pivot(index='hoten', columns='mon', values='diem')
df_pivot
```

```python
df_pivot.reset_index(inplace=True)
df_pivot.index.name = None
df_pivot.columns.name = None
df_pivot
```
