```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

### dùng replace để thay thế cột sex

```python
df['sex_1'] = df['sex'].str.replace('female', '0').str.replace('male', '1').astype(int)
df.head()
```

```python
df['sex_2'] = df['sex'].apply(lambda x: 0 if x == 'female' else 1)
df
```

```python
def bien_doi_gioi_tinh(value):
    if value=='female':
        return 0
    else:
        return 1
```

```python
df['sex_3'] = df['sex'].apply(lambda x: bien_doi_gioi_tinh(x))
df
```

```python
df['sex_4'] = df['sex'].apply(bien_doi_gioi_tinh)
df
```
