```python
import pandas as pd
import numpy as np
```

## TH 1

```python
df1 = pd.DataFrame({
    'hoten': ['nvb', 'ltc', 'nva', 'nvc'],
    'gioitinh': ['nam', 'nữ', 'nam', 'nam']
})
df1
```

```python
df2 = pd.DataFrame({
    'hoten': ['nva', 'nvb', 'ltc', 'ltd'],
    'tuoi': [25, 22, 30, 28]
})
df2
```

```python
# kết nối 2 bảng sử dụng phép kết inner (tương đương)
# mặc định: how='inner'
df_new1 = pd.merge(left=df1, right=df2, on=['hoten']) # hoten là cột chung
df_new1
```

```python
# kết nối 2 bảng sử dụng phép kết inner (tương đương) có sắp xếp lại dữ liệu
# mặc định: how='inner'
df_new1b = pd.merge(left=df1, right=df2, on=['hoten'], sort=True)
df_new1b
```

```python
# kết nối 2 bảng sử dụng phép kết left (ưu tiên cho bảng bên trái)
df_new1c = pd.merge(left=df1, right=df2, on=['hoten'], how='left') #lấy tất cả các dữ liệu (cả null) bên trái
df_new1c
```

```python
# kết nối 2 bảng sử dụng phép kết right (ưu tiên cho bảng bên phải)
df_new1d = pd.merge(left=df1, right=df2, on=['hoten'], how='right')
df_new1d
```

```python
# kết nối 2 bảng sử dụng phép kết outer (ưu tiên cho cả 2 bảng)
df_new1e = pd.merge(left=df1, right=df2, on=['hoten'], how='outer')
df_new1e
```

## TH 2

```python
df1 = pd.DataFrame({
    'hoten': ['nvb', 'ltc', 'nva', 'nvc'],
    'gioitinh': ['nam', 'nữ', 'nam', 'nam']
})
df1
```

```python
df2 = pd.DataFrame({
    'ten': ['nva', 'nvb', 'ltc', 'ltd'],
    'tuoi': [25, 22, 30, 28]
})
df2
```

```python
# kết nối 2 bảng sử dụng phép kết inner
# cột 1 là họ tên
# cột 2 là tên
# => có cột chung nhưng khác tên cột

df_new2 = pd.merge(left=df1, right=df2, left_on=['hoten'], right_on=['ten'])
df_new2
```

## TH 3

```python
df1 = pd.DataFrame({
    'hoten': ['nvb', 'ltc', 'nva', 'nvc'],
    'gioitinh': ['nam', 'nữ', 'nam', 'nam']
})
df1
```

```python
df2 = pd.DataFrame({
    'hoten': ['nva', 'nvb', 'ltc', 'ltd'],
    'tuoi': [25, 22, 30, 28]
})
df2.set_index('hoten', inplace=True)
df2
```

```python
# kết nối 2 bảng sử dụng phép kết inner
# df1: cột họ tên
# df2: cột họ tên làm index

df_new3 = pd.merge(left=df1, right=df2, left_on=['hoten'], right_index=True)
df_new3
```

```python
df1 = pd.DataFrame({
    'hoten': ['nvb', 'ltc', 'nva', 'nvc'],
    'gioitinh': ['nam', 'nữ', 'nam', 'nam']
})
df1.set_index('hoten', inplace=True)
df1
```

```python
df2 = pd.DataFrame({
    'ten': ['nva', 'nvb', 'ltc', 'ltd'],
    'tuoi': [25, 22, 30, 28]
})
df2.set_index('ten', inplace=True)
df2
```

```python
# kết nối 2 bảng sử dụng phép kết inner
'''
df1: cột 'họ tên' làm index
df2: cột 'tên' là index
'''
df_new4 = pd.merge(left=df1, right=df2, left_index=True, right_index=True)
df_new4
```
